```python
def create_sequence(data, seq_lenth=30):
    x, y = [], []
    for i in range(len(data) - seq_lenth):
        x.append(data[i:i+seq_lenth])
        y.append(data[i+seq_lenth])
        return numpy.array(x) , numpy.array(y)
```

```python
seq_lenth = 100
x, y = create_sequence(df_google, seq_lenth)
```

```python
train_size = 0.8

x_train = x[:int(len(x) * train_size)]
y_train = y[:int(len(y) * train_size)]

y_test = y[int(len(y) * train_size):]
x_test = x[int(len(x) * train_size):]
```

```python
x_train = torch.FloatTensor(x_train)
x_test = torch.FloatTensor(x_test)
y_train = torch.FloatTensor(y_train)
y_test = torch.FloatTensor(y_test)
```

```python
from zmq import device


class StockPriceLstm(nn.Module):
    def __init__(self, input_size, hidden_size=50, num_layer=2, output_size=1):
        super(StockPriceLstm, self).__init__()

        self.lstm = nn.LSTM(input_size, hidden_size, num_layer)
        self.linear = nn.Linear(hidden_size, output_size)
        self.num_layer = num_layer
        self.hidden_size = hidden_size

    def forward(self, x):
        h0 = torch.zeros(self.num_layer, x.shape[0], self.hidden_size).to(device, torch.FloatTensor)
        c0 = torch.zeros(self.num_layer, x.shape[0], self.hidden_size).to(device, torch.FloatTensor)

        out, _ = self.lstm(x, (h0, c0))
        out = self.linear(out[:, -1, :])

        return out
```

```python
lstm_model = StockPriceLstm(seq_lenth)
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.01)
```

```python
epochs = 1000
for epoch in range(epochs):
    y_pred = lstm_model(x_train)
    loss = loss_fn(y_pred, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f"Epoch: {epoch}, Loss: {loss.item()}")
```
