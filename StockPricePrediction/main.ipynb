```python
stock_data = torch.FloatTensor(df_google).view(-1)
```

```python
def create_sequence(data, seq_lenth=30):
    xs, ys = [], []
    for i in range(len(data) - seq_lenth):
        x = data[i:i+seq_lenth]
        y = data[i+seq_lenth]
        xs.append(x)
        ys.append(y)
    return torch.stack(xs), torch.stack(ys)
```

```python
seq_lenth = 100
x, y = create_sequence(stock_data, seq_lenth)
```

```python
train_size = int(len(y) * 0.8)

x_train = x[:train_size]
y_train = y[:train_size]

y_test = y[train_size:]
x_test = x[train_size:]
```

```python
from zmq import device


class StockPriceLstm(nn.Module):
    def __init__(self, input_size=1, hidden_size=50, num_layer=1, output_size=1, batch_first=True):
        super(StockPriceLstm, self).__init__()

        self.lstm = nn.LSTM(input_size, hidden_size, num_layer, batch_first=batch_first)
        self.linear = nn.Linear(hidden_size, output_size)
        self.num_layer = num_layer
        self.hidden_size = hidden_size

    def forward(self, x):
        h0 = torch.zeros(self.num_layer, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layer, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.lstm(x, (h0, c0))
        out = self.linear(out[:, -1, :])

        return out
```

```python
lstm_model = StockPriceLstm(input_size=seq_lenth)
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(lstm_model.parameters(), lr=0.01)
```

```python
epochs = 2000
for epoch in range(epochs):
    for x_batch, y_batch in zip(x_train, y_train):
        y_pred = lstm_model(x_train)
        loss = loss_fn(y_pred, y_train)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if epoch % 100 == 0:
            print(f"Epoch: {epoch}, Loss: {loss.item()}")
```
